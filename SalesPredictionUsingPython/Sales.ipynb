```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
xtrain=xtrain.astype(int)
ytrain=ytrain.astype(int)
xtest=xtest.astype(int)
ytest=ytest.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
xtrainscaled=sc.fit_transform(xtrain)
```

```python
xtestscaled=sc.fit_transform(xtest)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(xtrainscaled,ytrain)
```

```python
ypred=lr.predict(xtestscaled)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(ytest,ypred)
```
